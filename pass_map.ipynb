```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/mckayjohns/youtube-videos/main/data/messibetis.csv')
```

```python
df.info
```

```python
print(df['x'])
```

```python
df['x'] = df['x']*1.2
df['y'] = df['y']*0.8
```

```python
df['endX'] = df['endX']*1.2
df['endY'] = df['endY']*0.8
```

```python
fig,ax = plt.subplots(figsize=(13.5,8))
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')

pitch = Pitch(pitch_color='#22312b', line_color='white',
              stripe=False, pitch_type='statsbomb', half=False)  # optional stripes
pitch.draw(ax=ax, figsize=(16, 11))
plt.gca().invert_yaxis()

for x in range (len(df['x'])):
    if df['outcome'][x] == 'Successful':
        plt.plot((df['x'][x],df['endX'][x]), (df['y'][x],df['endY'][x]), color='green')
        plt.scatter(df['x'][x],df['y'][x], color='green')
    if df['outcome'][x] == 'Unsuccessful':
        plt.plot((df['x'][x],df['endX'][x]), (df['y'][x],df['endY'][x]), color='red')
        plt.scatter(df['x'][x],df['y'][x], color='red')

plt.title('Messi pass Map vs Real Betis')
```
